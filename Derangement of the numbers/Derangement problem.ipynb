A derangement of the numbers 1 through n is a permutation of all n those numbers such that none of them is in the “right place.” For example, 34251 is a derangement of 1 through 5, but 24351 is not because 3 is in the 3rd position.We will use simulation to estimate the number of derangements of the numbers 1 through 12.

Here is a program that generates random permutations of the integers 1, 2,…, 12. The program determines whether or not each permutation is a derangement.

```python
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng(1)
n = 1
a = pd.DataFrame(rng.integers(0,12, size = 12).reshape((1,12)))
while n < 1000:
    a.loc[len(a.index)] = rng.integers(0,12, size = 12)
    n = n+1

a.columns = np.arange(1,13)
a
```

```python
b = pd.DataFrame(np.arange(1,13).reshape(1,12))
b.columns = np.arange(1,13)
b

```

```python
c = pd.DataFrame(np.arange(1,12001).reshape(1000,12))
c.columns = np.arange(1,13)
c
```

```python
i = 0
j = 1
while i <= 1000 and j <= 12:
    if a.iloc[i][j] == b.iloc[0][j]:
        c.iloc[i][j] = 1
    else:
        c.iloc[i][j] = 0
    i = i+1
    if i == 1000:
        j = j+1
        i = 0

c

```

```python
c['der'] = c.sum(axis = 1)
x = len(c[c['der'] == 0])
x
print(f'Probability of randomly creating a derangement from 1 to 12 is ~{x*100/len(c)} %')
```
